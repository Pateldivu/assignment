A3] Write a Python program to sum of three givenintegers.
However, if two values are equal sum will be zero.

```python
def sum_three(x, y, z):
    if x == y or y == z or x == z:
        sum = 0
    else:
        sum = x + y + z
    return sum
print(sum_three(2, 1, 2))
print(sum_three(3, 2, 2))
print(sum_three(2, 2, 2))
```

I2 ] Write a Python program to add 'ing' at the end of a givenstring
(length should be at least 3). If the given string already ends with 'ing' then add 'ly' instead If the string length of the given string is less than 3, leave itunchanged

```python
def string(str1):
  length = len(str1)

  if length > 2:
    if str1[-3:] == 'ing':
      str1 += 'ly'
    else:
      str1 += 'ing'

  return str1
print(string('ja'))
print(string('jas'))
print(string('jasing'))
```

I4]  Write a Python function that takes a list of words and returns the length of the longestone.

```python
def find_longest_word(words_list):
    word_len = []
    for n in words_list:
        word_len.append((len(n), n))
    word_len.sort()
    return word_len[-1][0], word_len[-1][1]
result = find_longest_word(["abc", "defghi", "jklmnop"])
print("Longest word: " f'{result[1]}')
print("Length of the longest word: "f'{result[0]}')
```

I5] Write a Python function to reverses a string if it's length is a multiple of 4

```python
def reverse_string(str1):
    if len(str1) % 4 == 0:
       return ''.join(reversed(str1))
    return str1

print(reverse_string('abcd'))
print(reverse_string('divya'))
```

```python
def string_both_ends(str):
  if len(str) < 2:
    return ''

  return str[0:2] + str[-2:]

print(string_both_ends('w3resource'))
print(string_both_ends('w3'))
print(string_both_ends('w'))
```
